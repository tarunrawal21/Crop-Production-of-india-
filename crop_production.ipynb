**IMPORTING LIBRARY**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from warnings import filterwarnings
filterwarnings ('ignore')
```

**IMPORTING DATASET**

```python
cp = pd.read_csv('/content/Crop Production data.csv')
cp
```

**Determining size of dataset**

```python
cp.shape
```

**Determining colums**

```python
cp.info()
```

```python
cp.describe()
```

**Searcing for missing values and then removing them**



```python
cp.isnull().sum()
```

```python
dataset = cp.dropna()
```

```python
dataset.isnull().sum()
```

```python
dataset['State_Name'].unique()
```

```python
dataset['Season'].unique()
```

```python
dataset['Crop_Year'].unique()
```

```python
dataset['District_Name'].unique()
```

**Data Visualization**

```python
sns.lineplot(x = 'Crop_Year', y = 'Production', data = dataset)
plt.title('Production over the years')
plt.show()
```

```python
sns.barplot(x = 'Season', y = 'Production', data = dataset)
plt.title('Production per season')
plt.show()
```

```python
plt.figure(figsize=(25,10))
sns.barplot(x = 'State_Name', y = 'Production', data = dataset)
plt.title('Production per State', size = 15)
plt.xticks(rotation=90)
plt.show
```

```python
sns.countplot(x = 'Season', data = dataset)
plt.title('Season')
plt.show()
```

```python
sns.pairplot(dataset, hue = 'Season')
plt.show()
```

```python
width = 10
height = 10
plt.figure(figsize=(width, height))
sns.barplot(x= 'Crop_Year', y = 'Production', data = dataset)
plt.title('Crop year and production', size = 25)
plt.show()
```

```python
width = 10
height = 10
plt.figure(figsize=(width, height))
sns.barplot(x= 'Crop_Year', y = 'Area', data = dataset)
plt.title('Area of production over the year', size = 25)
```
